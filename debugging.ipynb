```python
# Keep headers that will go in gii files
metaFake_Left = {'AnatomicalStructurePrimary': 'CortexLeft', 
    'PaletteNormalizationMode': 'NORMALIZATION_SELECTED_MAP_DATA', 
    'TimeStep': '1.3', 
    }
metaFake_Right = {'AnatomicalStructurePrimary': 'CortexRight',
    'PaletteNormalizationMode': 'NORMALIZATION_SELECTED_MAP_DATA',
    'TimeStep': '1.3',
    }
giiMeta_Left = nib.gifti.GiftiMetaData( metaFake_Left )
giiMeta_Right = nib.gifti.GiftiMetaData( metaFake_Right )
```

```python
fitEstims = NiiFlatten.astype(np.float32)
imgLeft = nib.gifti.GiftiImage(meta=giiMeta_Left)
imgRight = nib.gifti.GiftiImage(meta=giiMeta_Right)
for i in range(fitEstims.shape[1]):
    dataNew_L = nib.gifti.GiftiDataArray(data = fitEstims[:numVertices_L, i])
    imgLeft.add_gifti_data_array(dataNew_L)
    dataNew_R = nib.gifti.GiftiDataArray(data = fitEstims[numVertices_L:numVertices_L+numVertices_R, i])
    imgRight.add_gifti_data_array(dataNew_R)
# for i in range(0, numVertices_L):
#     dataNew_L = nib.gifti.GiftiDataArray(data = fitEstims[i, :])
#     imgLeft.add_gifti_data_array(dataNew_L)
# for i in range(0, numVertices_R):
#     dataNew_R = nib.gifti.GiftiDataArray(data = fitEstims[i+numVertices_L, :].T)
#     imgRight.add_gifti_data_array(dataNew_R)
```
